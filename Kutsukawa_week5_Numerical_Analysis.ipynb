### Analysis of Single Numeric Variable

#### Customer_Age

````python
#### Count Customer_Age summary statistics
pdf['Customer_Age'].describe()
````

There are fewer customers below 30 and above 60 years old compared to the middle age groups. This suggests that customer acquisition might be more focused on middle-aged.

````python
#### Get Customer_Age boxplot
pdf['Customer_Age'].plot(kind='box')
````

The max percentile is below 70; the lowest percentile is below 30. Most customers' ages fall between 41 and 52, and there's only one outlier, over 70 years of age.

````python
#### Get Dependent_count histogram
pdf['Dependent_count'].plot(kind='hist')
````

The max percentile is 5 dependent count and the lowest percentile is 0 dependent count is 0. Most of the dependent count falls between 1 to 3. There's no outlier in the box plot.

### Analysis of Two Numeric Variables

#### Customer_Age vs Dependent_count

````python
#### Get Dependent_count vs Customer_Age correlation
pdf[['Dependent_count', 'Customer_Age']].corr()
````

````python
#### Show Dependent_count vs Customer_Age scatter plot
pdf.plot(kind='scatter', x='Dependent_count', y='Customer_Age')
````

Based on the scatter plot for that 0 dependent_count, you can see the range of the customer age ranges from below 30 and above 70, but as dependent_count gets higher, the range of the customer age gets younger in their average, so as you can see in the 5 dependent_count the range only falls onto below 40 and below 60.

#### Months_Inactive_12_mon vs Credit_Limit

````python
#### Get Months_Inactive_12_mon vs Credit_Limit correlation
pdf[['Months_Inactive_12_mon', 'Credit_Limit']].corr()
````

````python
#### Show Months_Inactive_12_mon vs Credit_Limit scatter plot
pdf.plot(kind='scatter', x='Months_Inactive_12_mon', y='Credit_Limit')
````

Based on the scatter plot, you can see that most customers become inactive in months 1, 2, and 3. You can see that all of the customer's credit limits become inactive during that month, but in months 4, 5, and 6, a few customers only become inactive in those months, and most of the credit limits fall below 10,000.
